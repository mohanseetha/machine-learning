<h1>IRIS Data Prediction</h1>

<h4>Importing neccessary libraries</h4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

<h4>Reading the data</h4>

```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
df.head()
```

```python
df.drop('Id', axis=1, inplace=True)
df.head()
```

<h4>Preprocessing the dataframe by converting categorical data to numerical data</h4>

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Species"] = le.fit_transform(df["Species"])
```

<h4>Understanding the correlation between the features of the dataset</h4>

```python
cor = df.corr()
sns.heatmap(cor, cmap="Blues", annot=True);
```

<h4>Bivariate analysis on Sepal Length and Width</h4>

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df["SepalLengthCm"], df["SepalWidthCm"])
ax.set_title("Sepal Length vs Width")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width");
```

<h4>Bivariate analysis on Petal Length and Width</h4>

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df["PetalLengthCm"], df["PetalWidthCm"])
ax.set_title("Petal Length vs Width")
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width");
```

<h4>Importing necessary Machine Learning libraries</h4>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

<h4>Splitting the data into training and testing data</h4>

```python
X = df.drop("Species", axis=1)
y = df["Species"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h4>Logisitic Regression</h4>

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_test, y_pred_lr)
ax.set_title('Real Value vs Predicted Value (Logistic Regression)')
ax.set_xlabel('Real Value')
ax.set_ylabel('Predicted Value');
```

```python
acc = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy Score for Logistic Regression: {acc*100:.2f}%")
cm = confusion_matrix(y_test, y_pred_lr)
print("Confusion Matrix:")
sns.heatmap(cm, cmap="Blues", annot=True);
```

<h4>Support Vector Machine</h4>

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_test, y_pred_svm)
ax.set_title('Real Value vs Predicted Value (Support Vector Machine)')
ax.set_xlabel('Real Value')
ax.set_ylabel('Predicted Value');
```

```python
acc = accuracy_score(y_test, y_pred_svm)
print(f"Accuracy Score for SVM: {acc*100:.2f}%")
cm = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix:")
sns.heatmap(cm, cmap="Blues", annot=True);
```

<h4>KNN</h4>

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_test, y_pred_knn)
ax.set_title('Real Value vs Predicted Value (KNN)')
ax.set_xlabel('Real Value')
ax.set_ylabel('Predicted Value');
```

```python
acc = accuracy_score(y_test, y_pred_knn)
print(f"Accuracy Score for KNN: {acc*100:.2f}%")
cm = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:")
sns.heatmap(cm, cmap="Blues", annot=True);
```

<h4>Decision Tree Classifier</h4>

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_test, y_pred_dt)
ax.set_title('Real Value vs Predicted Value (Decision Tree)')
ax.set_xlabel('Real Value')
ax.set_ylabel('Predicted Value');
```

```python
acc = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy Score for Decision Tree: {acc*100:.2f}%")
cm = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix:")
sns.heatmap(cm, cmap="Blues", annot=True);
```
