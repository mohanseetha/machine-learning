```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = df.drop(['Engine Fuel Type', 'Number of Doors', 'Driven_Wheels', 'Market Category', 'Vehicle Size', 'Vehicle Style', 'Popularity'], axis=1)
df.head()
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
df.head()
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.head()
```

```python
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(IQR)
```

```python
numeric_df = numeric_df[~((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)]
numeric_df.shape
```

```python
remaining_indices = numeric_df.index
df = df.loc[remaining_indices].reset_index(drop=True)
df.shape
```

```python
df['Make'].value_counts().plot(kind='bar', figsize=(10, 5))
plt.title('Number of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Number of Cars');
```

```python
plt.figure(figsize=(10, 5))
cor = numeric_df.corr()
sns.heatmap(cor, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.yticks(rotation=0);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df['HP'], df['Price'])
ax.set_title('HP vs Price')
ax.set_xlabel('HP')
ax.set_ylabel('Price');
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.regplot(x='HP', y='MPG-H', data=df, ax=ax)
ax.set_title('Regression Plot')
ax.set_xlabel('HP')
ax.set_ylabel('Price');
```
